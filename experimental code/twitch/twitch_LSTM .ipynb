```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size):
        super().__init__()
        
        self.hidden_size = hidden_size
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.it_embedding = nn.Embedding(16440, self.hidden_size)
        self.du_embedding = nn.Embedding(95, self.hidden_size)
        self.int_embedding = nn.Embedding(21, self.hidden_size)
        self.lstm = nn.LSTM(hidden_size*3, hidden_size*3, batch_first=True)
        self.out_it = nn.Linear(self.hidden_size*3, 16440)
        self.out_du = nn.Linear(self.hidden_size*3, 95)
        self.out_int = nn.Linear(self.hidden_size*3, 21)
        
        
    def forward(self, input):
        
        items = input['item'] 
        target_items = input['target_item']
        durations = input['duration'] 
        target_durations = input['target_duration']
        intervals = input['interval'] 
        target_intervals = input['target_interval'] 
        
        # 轉換輸入為 embedding
        embedded_item = self.it_embedding(items)
   
        embedded_duration = self.du_embedding(durations)
        embedded_interval = self.int_embedding(intervals)
        embedded = torch.cat((embedded_item,embedded_duration,embedded_interval),dim=2)
 
        output, (h_n, c_n) = self.lstm(embedded)
        output = (output[:, -1, :])
        output1 = self.out_it(output)
        output2 = self.out_du(output)
        output3 = self.out_int(output)
        return output1,output2,output3

      
```
